```python
baseurl='https://services.cancerimagingarchive.net/nbia-api/services/v1/'
def restCall(url,itemName):
  response = requests.get(baseurl+url)
  if len(response.text)==0:
    return []
  # retList = set(item[itemName] for item in response.json())
  retList = [] 
  for d in response.json():
    if itemName in d:
      retList.append(d[itemName])
  return retList
```

```python
## creates a TCIA file with the provided series list. 
def createTCIAmanifest(outputFileName,listOfSeriesToDownload,databasketId='manifest-1646842161649.tcia'):
  """  
    outputFileName: outputFileName
    databasketId: dataID needed by TCIA to download from 
    listOfSeriesToDownload:  listOfSeriesToDownload
  """
  manifestHeader = ['downloadServerUrl=https://public.cancerimagingarchive.net/nbia-download/servlet/DownloadServlet', 
                  'includeAnnotation=true',
                  'noOfrRetry=4',
                  'manifestVersion=3.0']
  manifestHeader.extend(['databasketId='+databasketId])
  manifestHeader.extend(['ListOfSeriesToDownload='])
  manifestHeader.extend(listOfSeriesToDownload)
  with open(outputFileName, 'w') as f:
    for line in manifestHeader:
      f.write(line)
      f.write('\n')
```

```python
# Goal is to only download the dcm that were segmented.
import pydicom
import os 
```

```python
##############################################  
def openDCM(dcmPath):
  tag4Type=(0x0008,0x0060) # to find type, could also use ,(0x0002, 0x0002),(0x0002, 0x0012)]
  tag4SEGreferenceSeries=[(0x0008, 0x1115)] # worked for CT images 
  tag4SEGreferenceSeries +=[(0x0008,0x1140)]  # ReferencedImageSequence needed when ref series is not populated
  tag4RTreferenceSeries =[(0x3006, 0x0010)]  
  tag4StudyID=[(0x0020,0x000D)] #StudyInstanceUID
  taglist= [tag4Type] + tag4SEGreferenceSeries+tag4RTreferenceSeries +tag4StudyID
  ds = pydicom.read_file(dcmPath,stop_before_pixels=True,specific_tags=taglist)
  return ds,ds[tag4Type].value
##############################################
def dcmSafeKeyAccess(ds,findfirstRefInst=False):
  look4keys=[(0x0020, 0x000e)]  #    nested ref SeriesInstanceUID	# worked for CT images 
  if findfirstRefInst:
    look4keys+=[(0x0008, 0x1155)]  # Referenced SOP Instance UID 
  returnStr=""
  for elem in ds:
    # print(f"elem.tag is {elem.tag} , elem.VR = {elem.VR}")
    if elem.VR == "SQ":
      for item in elem:
        returnStr=dcmSafeKeyAccess(item,findfirstRefInst)
    if elem.tag in look4keys:
      # print(f"ref series is {str(elem)}")
      return elem.value
  return returnStr
##############################################
def match_refInstance_w_each_series_in_Study(StudyInstanceUID,refInstID):
  # doing work around when ref Series UUID is not populated 
  # get all sereies in the study and compare it with 1 of the ref series ID in the seg/rt struct
  # print(f" refInstID {refInstID} ,series lst len ={len(seriesLst)}")
  seriesLst=restCall('getSeries?StudyInstanceUID='+StudyInstanceUID,'SeriesInstanceUID')
  for s,seriesID in enumerate(seriesLst):
    instLst=restCall('getSOPInstanceUIDs?SeriesInstanceUID='+seriesID,'SOPInstanceUID')
    # print(f" inst {s} len = {len(instLst)}")
    for i,instID in enumerate(instLst):
      if instID==refInstID:
        return seriesID
  return "" # failed 
##############################################
def getRefUUIDs(ROOT_FLD_DCM):
  refUUIDList = []
  for root, subFolders, files in os.walk(ROOT_FLD_DCM):
    for fileName in files:
      file = os.path.join(root,fileName)
      parentDirPath=os.path.dirname(file)
      parentDirName=os.path.basename(os.path.normpath(parentDirPath))
      ext=os.path.splitext(file)[1]
      if ext.lower() != ".dcm":
        continue
      ds,modality=openDCM(dcmPath=file)
      
      if modality in ["SEG","RTSTRUCT"]:
        refUUID = dcmSafeKeyAccess(ds=ds)
        if refUUID =="": # DICOM seg is not implemented correctlying standard ref Series tag is empty 
          print("---DICOM seg is not implemented correctlying standard ref Series tag is empty")
          # need to do more work to find referenced series ID
          refInstID = dcmSafeKeyAccess(ds=ds,findfirstRefInst=True)
          refUUID=match_refInstance_w_each_series_in_Study(ds.StudyInstanceUID,refInstID)
        print(f"file name is {fileName} parent is {parentDirName} , modality = {modality} refUUID is {refUUID}")
        refUUIDList.append(refUUID)
  return refUUIDList
```

```python
import nibabel as nib
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
import numpy as np
import pydicom_seg
import SimpleITK as sitk
```

```python
if True:  # you don't know the names of the rtstruct 
  nameLbIndMap={}                 
else: ### if you know rt structs names and want to map them to certain labels
  nameLbIndMap={"Lung_L":2,"Lung_R":3,"Heart":4,"Esophagus":5
              ,"Bones":6,"Spinal Canal":7, "Femoral Head Lef":8,"Femoral Head Rig":9
              ,"Kidney Left":11,"Kidney Right":12,"Liver":13,"Spleen":14
              ,"Stomach":15,"Duodenum":16
              ,"Pancreas":17
              ,"Gall Bladder": 18,"Small Intestine":19,"Large Intestine":20
              ,"Bladder":21,"Rectum":22,"Prostate":23
              ,"Gonads":31,"Thymus":32,"Adrenal Left":33,"Adrenal Right":34
              ,"UteroCervix":35,"Breast Left":36,"Breast Right":37
              }
#############  
def search4DCMImgDir(dirPath,subfoldersList):
  # TODO need better way to look for DCM dir other than matching part of the name
  for fld in subfoldersList:
    if "-CT-" in fld or "-arterial-"in fld:
      return os.path.join(dirPath,fld)
#############
def addRTname2Dict(segDirPath,nameLbIndMap):  
  rtFiles = [f for f in os.listdir(segDirPath)]
  for rtFile in rtFiles:
    if rtFile.startswith("_all_"):
      continue
    rtFile=rtFile[:rtFile.find(".nii.gz")]
    # print (rtFile)
    if rtFile not in nameLbIndMap.keys():
      nameLbIndMap[rtFile]=len(nameLbIndMap)+1
#############
def mergeNlabel(segDirPath,outputPath,nameLbIndMap=nameLbIndMap):
  print(f"segDirPath {segDirPath}")
  addRTname2Dict(segDirPath,nameLbIndMap)
  # print(nameLbIndMap)
  first=True
  for fname,lb in nameLbIndMap.items():
    segPath=segDirPath+fname+".nii.gz"
    if not os.path.exists(segPath):
      continue
    vol = nib.load(segPath)
    vol_data = vol.get_fdata()
    if first:
      all = np.zeros_like(vol_data)
      first = False
    all[vol_data>0]=lb
  nib.Nifti1Image(all,vol.affine).to_filename(outputPath)      
#############  
def loop_over_Dataset(convertModality="CT"): #"CT" "RTSTRUCT" or "SEG"
  for root, subFolders, files in os.walk(ROOT_FLD_DCM):
    if len(subFolders)>1:
      refDCMPath=search4DCMImgDir(root,subFolders)
    for fileName in files:
      file = os.path.join(root,fileName)
      ext=os.path.splitext(file)[1]
      if ext.lower() != ".dcm":
        continue
      parentDirPath=os.path.dirname(file)
      parentDirName=os.path.basename(os.path.normpath(parentDirPath))
      patientDirPath=os.path.basename(os.path.normpath(os.path.dirname(os.path.dirname(parentDirPath))))
      patientDirName=os.path.basename(patientDirPath)
      if APPEND_SERIES_NAME2PATIENT:
        # patientDirName += "/"+ parentDirName.replace(" ","")
        patientDirName += "/"+ parentDirName.replace(" ","")[-10:] # series discribtion can be very long name, use only last 8 char
      ds,modality=openDCM(dcmPath=file)
      if (modality == "CT" and convertModality=="CT") or (modality == "MR" and convertModality=="MR"):
        print(f"file name is {fileName} patientDirName is {patientDirName} , modality = {modality} dcmDir= {parentDirPath}")
        niiPath=NII_FOLDER+patientDirName+"_img.nii.gz"
        !plastimatch convert --input '$parentDirPath' --output-img '$niiPath'
        break
      elif modality == "SEG" and convertModality=="SEG":
        print(f"file name is {fileName} patientDirName is {patientDirName} , modality = {modality} --> convert SEG  dcmPath= {refDCMPath} ")
        dcmSeg2Nii(file,NII_FOLDER+patientDirName+"_seg.nii.gz")
      elif modality == "RTSTRUCT" and convertModality=="RTSTRUCT":
        print(f"file name is {fileName} patientDirName is {patientDirName} , modality = {modality} --> convert RT Struct  dcmPath= {refDCMPath}")

        # there could be multiple RT dcm files for the same study as 1-1.dcm and 1-2.dcm --> adding 3rd char from file name
        pat_temp_dir=NII_FOLDER_SINGLE_LB+patientDirName+"/"+fileName[2]+"/"
        segNii=pat_temp_dir+"_all_RTseg.nii.gz"      
        outputMap=pat_temp_dir+"map"+".txt"        
        !plastimatch convert --referenced-ct '$refDCMPath' --input '$file' --output-labelmap '$segNii' --output-prefix '$pat_temp_dir' --prefix-format "nii.gz" 
        ###!plastimatch convert --referenced-ct '$refDCMPath' --input '$file' --output-labelmap '$segNii' --output-colormap '$outputMap'
        mergeNlabel(pat_temp_dir,NII_FOLDER+patientDirName+"_RTseg.nii.gz",nameLbIndMap)
        # !plastimatch convert --referenced-ct '$refDCMPath' --input '$file' --output-img '$segNii' --prefix-format "nii.gz"
#############
def dcmSeg2Nii(dcmSegPath, outputNiiPath="",refImgNiiPath="", flipAxis=""):
  if outputNiiPath == "":
    outputNiiPath = os.path.dirname(dcmSegPath) + "/" + os.path.basename(dcmSegPath).replace(".dcm", ".nii.gz")
    print(" empty dir output to same dir=", outputNiiPath)
  print(dcmSegPath)
  dcm = pydicom.dcmread(dcmSegPath)
  reader = pydicom_seg.SegmentReader()
  result = reader.read(dcm)

  print(" dicom seg has {} labels ".format(result.available_segments))
  allNp = None
  for segment_number in result.available_segments:
    # print("going through segments", segment_number)
    # single = result.segment_data(segment_number)  # directly available
    image = result.segment_image(segment_number)  # lazy construction
    single = sitk.GetArrayFromImage(image)
    print("non zero count for ", segment_number, " is=", np.count_nonzero(single))
    if allNp is None:
        allNp = np.zeros_like(single, dtype=np.uint8)
    single[allNp > 0] = 0  # Xnat has a strange saving where labels write over each others
    # Need to clear first then set that segment
    allNp[single > 0] = segment_number

  print("GetDirection =", image.GetDirection(), "  spaceing =========", image.GetSpacing(),
        "GetOrigin =", image.GetOrigin(), "  GetSize =", image.GetSize(), "   np shape=", allNp.shape)

  outputNiiPath = outputNiiPath.replace(".nii.gz", "")
  sitkSegmentation = sitk.GetImageFromArray(allNp)
  sitkSegmentation.CopyInformation(image)
  sitk.WriteImage(sitkSegmentation, outputNiiPath + ".nii.gz", True)
  print(f"------------- worte file to {outputNiiPath}.nii.gz")

```

```python
def showSampleSlices(vol_data,seg_data=None,fig_rows = 4,fig_cols = 4 ,figsize=[20, 20]):
  print(f"{vol_data.shape} mix is {np.amin(vol_data)} max is {np.amax(vol_data)}")

  n_subplots = fig_rows * fig_cols
  n_slice = vol_data.shape[2]
  step_size = n_slice // n_subplots
  plot_range = n_subplots * step_size
  start_stop = int((n_slice - plot_range) / 2)
  if seg_data is None:
    fig, axs = plt.subplots(fig_rows, fig_cols, figsize=figsize)
  else:
    print(f"{seg_data.shape} mix is {np.amin(seg_data)} max is {np.amax(seg_data)}")
    fig_cols =int(fig_cols/2)
    fig_rows =fig_rows *2
    fig, axs = plt.subplots(fig_rows, fig_cols*2, figsize=figsize)
  
  if seg_data is None:
    for idx, img in enumerate(range(start_stop, plot_range, step_size)):
      axs.flat[idx].imshow(ndi.rotate(vol_data[:, :,img], 90), cmap='gray')
      axs.flat[idx].axis('off')
  else:
    for idx, img in enumerate(range(start_stop, plot_range, step_size)):
      axs.flat[idx*2].imshow(ndi.rotate(vol_data[:, :,img], 90), cmap='gray')
      axs.flat[idx*2].axis('off')
      axs.flat[idx*2+1].imshow(ndi.rotate(seg_data[:, :,img], 90) )
      axs.flat[idx*2+1].axis('off')

  plt.tight_layout()
  plt.show()

def openNshow(imagePath,segPath=None):
  #open sample file
  vol = nib.load(imagePath)
  vol_data = vol.get_fdata()
  if segPath is not None:
    seg = nib.load(segPath)
    seg_data = seg.get_fdata()
  else:
    seg_data=None
  showSampleSlices(vol_data,seg_data)  

```
